TASK 1 - DATA IMMERSION & WRANGLING (ApexPlanet Internship)


```python
import pandas as pd 
import numpy as np 

# Load Dataset
df=pd.read_excel(r"c:\Users\dell\Downloads\ApexPlanet_DataAnalytics_Dataset.xlsx")
```

```python
# DATA IMMERSION
print("Dataset shape:",df.shape)

print("\nColumns:",df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Records:")
print(df.head())
```

```python
# DATA QUALITY ASSESSMENT
print("Missing values")
print(df.isnull().sum())
```

```python
print("Duplicate Rows:",df.duplicated().sum())

print("\nDuplicate Order IDs:",df["Order_ID"].duplicated().sum())
```

DATA CLEANING

```python
# Remove duplicate Order IDs
df=df.drop_duplicates("Order_ID")
```

```python
# Fill Missing Age with Median
df["Age"]=df["Age"].fillna(df["Age"].median())
```

```python
# Fill Missing City with Mode
df["City"]=df["City"].fillna(df["City"].mode()[0])
```

```python
# Convert Date Column
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
```

```python
# Remove Extra Spaces
df["Customer_Name"]=df["Customer_Name"].str.strip()
```

```python
# Standardize Gender
df["Gender"]=df["Gender"].str.title()
```

DATA TRANSFORMATION

```python
# Create Month Column
df["Month"]=df["Order_Date"].dt.month_name()
```

```python
# Create Sales Category
df["Sales_Category"]=np.where(df["Total_Sales"]>100000,"High Sales","Low Sales")
```

```python
# Create Age Group
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18,30,45,60,100],
    labels=['Young','Adult','Middle Age','Senior'],
    include_lowest=True
)
```

FINAL CHECK


```python
print("Final Shape:",df.shape)
```

```python
print("Remaining Missing Values:",df.isnull().sum())
```

```python
# EXPORT CLEAN DATA
df.to_csv("Cleaned_Dataset.csv", index=False)
```
